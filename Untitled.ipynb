***
  # Maxwell distribution
***

Consider a gas in thermal equilibrium composed of one or more species $s_{i}$ (electrons, protons, ions). Each specie $i$ of mass $m_{i}$, has particles moving in all directions at different speeds $v$. Therefore, each specie has its own distribution function. The most probable distribution function of these velocities is given by the Maxwell distribution function. In one dimension, the Maxwell distribution function of the specie $i$ is given by

$$ f_{i}(v_{x}) = \displaystyle A\exp\Big(-\frac{m_{i}v_{x}^{2}}{2k_{B}T}\Big),$$

where $A$ is a constant, $T$ is the average temperature of the particles, $k_{b} = 1.38\times 10^{-23}$ [JK$^{-1}$] is the Boltzmann constant.

For a normalized Maxwell distribution function, this is 

$$ \int_{\mathcal{R}} f(v_{x}) = 1, $$ 

the constant $A$ is 

$$ A = \Big(\frac{m}{2\pi k_{b}T}\Big)^{1/2}.$$


Thus, a normalized 1D Maxwell distribution function takes the form

$$ f(v_{x}) =\Big(\frac{m}{2\pi k_{b}T}\Big)^{1/2} \exp\Big(-\frac{mv_{x}^{2}}{2k_{B}T}\Big),$$
&emsp;

Defining the thermal velocity as 

$$v_{th} = \Big( \frac{2k_{b}T}{m} \Big)^{1/2},$$

the distribution function can be rewriten as

$$f(v_{x}) = \frac{1}{\sqrt\pi v_{th}} \exp\Big(-\frac{v_{x}^{2}}{v_{th}^{2}}\Big)$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,6)

def Maxwell_1D(x, u):
    f = np.sqrt(1/(np.pi*u))*np.exp(-x**2)
    return f

u = np.linspace(-10, 10, 1000)

plt.figure()
plt.plot(u, Maxwell_1D(u, 1), label = r"$v_{th}$ = 1", linewidth = 1.5)
plt.plot(u, Maxwell_1D(u, 5), label = r"$v_{th}$ = 5", linewidth = 1.5)
plt.plot(u, Maxwell_1D(u, 10), label = r"$v_{th}$ = 10", linewidth = 1.5)
plt.xlabel(r"$v_{x}$", fontsize = "13")
plt.ylabel(r"f$(v_{x})$", fontsize = "14")
plt.grid(which='both', axis='both')
plt.tick_params(axis = "both", direction= "in", length = 5, width = 1, colors = "k", grid_alpha = 1)
plt.legend(loc = "best",  ncol = 1, fancybox=True, columnspacing = 1)
plt.xlim(3, -3)
plt.tight_layout()
plt.show()
```

Extending this result to three dimensions, the Maxwell distribution function is given by

$$ f(v_{x}, v_{y}, v_{z}) = \Big(\frac{m}{2\pi k_{b}T}\Big)^{3/2} \exp\Bigg(-\frac{m(v_{x}^{2}+v_{y}^{2}+v_{z}^{2})}{2k_{B}T}\Bigg). $$
